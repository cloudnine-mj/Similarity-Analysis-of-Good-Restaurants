```python
df = pd.read_csv('소상공인시장진흥공단_상가(상권)정보_부산_202103 - 복사본.csv', encoding='cp949')

df
```

```python
# 음식점 데이터만 쓰기 위함
df = df.loc[df['상권업종대분류명'] == '음식']  

# 필요한 칼럼만 골라냄
df  = df [['상호명', '상권업종중분류명', '상권업종소분류명', '표준산업분류명', '행정동명', '위도', '경도']]
```

```python
df.columns = ['Name',   # 상호명
             '중분류명',  # 상권업종중분류명
             '소분류명',  # 상권업종소분류명
             '표준산업분류명',  # 표준산업분류명
              'dong',   # 행정동 명
              'lat',    # 위도
              'lon']    # 경도
df
```

```python
df
```

```python
df['표준산업분류명'].isnull().sum()
```

```python
df['표준산업분류명'] = df['표준산업분류명'].fillna('한식 음식점업')
```

```python
df['Column_mix'] = df['중분류명'] + df['소분류명'] + df['표준산업분류명']
df['Column_mix'] = df['Column_mix'].str.replace("/", " ")
```

```python
df
```

```python
#dong별 점포수량 집계

group_base = df.groupby('dong')  #자치구별로 데이터 값을 묶어 집계
group_gu = group_base.size() # 배열의 모양과 크기 확인
df_gu = pd.DataFrame(group_gu, columns=['음식점갯수'])

df_gu.sort_values(by=['음식점갯수'],ascending = False)
```
